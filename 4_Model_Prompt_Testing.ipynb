# 跨模型基準測試: 比較不同 LLM 在特定主題下的表現

### <font color='cornflowerblue'>摘要：</font>



在此份作業中，將選擇了兩種以上的大型語言模型 LLM，針對特定主題設計一系列基準測試 prompts ，並比較它們的回答內容。

將比對不同 LLM 的回答之準確性、清晰度、深度與邏輯性上的差異，最後紀錄下觀察與結果。

## <font color='lightsalmon'>1. 選定主題</font>

##「超現實主義」
作為我在這次 LLM 基準測試的主題，原因是這個藝術流派讓我留下了深刻的印象。
之前參觀過薩爾瓦多·達利（Salvador Dalí）的畫展，透過他的作品，能夠體會到超現實主義獨特的風格。
包含充滿夢境般的意象、強烈的情感張力，以及對現實與潛意識的探索。
這個領域不僅涉及藝術，還牽涉了文學、哲學與心理學等，我認為適合作為測試 LLM 知識廣度與表達能力的主題。
觀察透過比較，不同的語言模型如何解讀超現實主義的核心概念或是對現代文化的影響。

#### <font color='#76BEE2'>ChatGPT 對於此句話的解釋偏向先知道答案，再回推該句話為什麼這樣說並解讀。</font>

## <font color='lightsalmon'>5. 結果比較統整</font>

### 1. <font color='indianred'>回答的精準度</font>

相比之下，Llama 3.3 的回答比較能直接切入重點，精準回應我的問題，不會額外加入太多的背景知識。
相較於 GPT 有時會提供過多細節與補充，感覺有些稀釋掉原本的重點。

### 2.  <font color='indianred'>表達方式與語言風格</font>

我認為 Llama3.3 的回答較為直接、簡單明瞭，給我一種貼近日常對話的感覺，容易閱讀與理解。
而 GPT 的回答我認為適合學術性較強或是專業領域的問題。

### 3.  <font color='indianred'>結論</font>

綜合來看，這次基準測試結果，我更喜歡 Llama 3.3 的回答方式，因為它能夠更精準地聚焦在問題，並以清晰、簡潔的方式回答。

但是這樣不表示 Llama 3.3 一定優於 GPT，而是要以使用者的需求與場景來評斷。
