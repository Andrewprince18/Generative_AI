# 畫Beta函數圖形

以下將畫出隨機變數 $X$ 服從 $Beta$ 分配，參數以 $a,b$ 為例，機率密度函數為以下

$f_x(x|a,b) = \dfrac{Γ(a+b)}{Γ(a)Γ(b)}x^{a-1}(1-x)^{b-1} \space ; \space 0<x<1 , \space a>0, \space b>0$

##函數介紹

貝它分配通常用於描述成功與失敗的比率，

其中兩個參數 $a,b$ 控制著分配圖形的形狀。當 $a=b$ 時圖形為對稱，因爲此時成功與失敗的機率各半。

1.   對稱性 :
     - 由於 $a=b$ ，機率密度函數是對稱的，中心位於 $x=0.5$ 。
     - 分佈形狀在 $[0,1]$ 區間內左右對稱
2.   期望值、變異數 :
     - $E[X]= \dfrac{a}{a+b} = \dfrac{1}{2}$
     - $Var(X) = \dfrac{ab}{(a+b)^2 \space (a+b+1)}$ ; 隨著 $a$ 增加，變異數變小，數值更集中於 0.5。
3.   圖形特徵 :
     - $a=b=1$ 時，分佈為均勻分佈，即 $f(x)=1$，表示隨機變數均勻地分佈在 $[0,1]$ 上。
     - $a=b>1$ 時，分佈開始在 $x=0.5$ 形成峰值，越大的 $a$ 分佈越集中於 $0.5$ 附近。
     - $a=b=2$ 時，形狀為拋物線。
     - $a=b$ 較大時，分佈圖形變得尖銳，表示幾乎所有值都集中在 0.5 附近，類似於常態分佈。


```python
from scipy.stats import beta

x = np.linspace(0, 1, 100)

plt.figure(figsize=(10, 6))

for a in range(1, 21):
    y = beta.pdf(x, a, a)

    plt.plot(x, y, label=f'a = b = {a}', color=plt.cm.Blues(a/20))

plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Beta Distribution for Various a=b')
plt.legend()
```
